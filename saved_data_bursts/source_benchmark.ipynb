```python
import matplotlib.pyplot as plt
```

```python
#create scatter plot in matplotlib with axes labels brightness and repetition rate

qd_warburton=[0.79,76]
qd_pan=[0.34, 76]

spdc_pan=[0.05, 80]
spdc_multi=[0.756, 0.5]

rb_korea=[0.17,10]
calcium=[0.08,0.1]
rb_rempe=[0.6, 1]
ca=[0.88,0.0024]

rb_us=[0.82,2]

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Computer Modern",
})

f, (a1) = plt.subplots(1, 1, sharex=True, figsize=(14, 7))
a1.plot(qd_warburton[0], qd_warburton[1], 'o',label='Quantum Dot', color='red', markersize=12)
a1.annotate('QD, 2021', (1.02*qd_warburton[0], qd_warburton[1]), fontsize=20)
a1.plot(qd_pan[0], qd_pan[1], 'o', color='red', markersize=12)
a1.annotate('QD *,  2019', (1.02*qd_pan[0], qd_pan[1]), fontsize=20)
a1.plot(spdc_pan[0], spdc_pan[1], 'o', color='black', markersize=12)
a1.annotate('SPDC,  2018', (1.2*spdc_pan[0], spdc_pan[1]), fontsize=20)
a1.plot(spdc_multi[0], spdc_multi[1], 'o',label='SPDC', color='black', markersize=12)
a1.annotate('SPDC, mplx  2019', (1.02*spdc_multi[0], spdc_multi[1]), fontsize=20)
a1.plot(rb_korea[0], rb_korea[1], 'o', color='blue', markersize=12)
a1.annotate('$^{85}\mathrm{Rb}$,  2011', (1.04*rb_korea[0], rb_korea[1]), fontsize=20)
a1.plot(ca[0], ca[1], 'o',label='Atom', color='blue', markersize=12)
a1.annotate('$\mathrm{Ca}^+$,  2011', (0.88*ca[0], ca[1]), fontsize=20)
a1.plot(rb_rempe[0], rb_rempe[1], 'o', color='blue', markersize=12)
a1.annotate('$^{87}\mathrm{Rb}$  2011', (1.02*rb_rempe[0], rb_rempe[1]), fontsize=20)
a1.plot(rb_us[0], rb_us[1], 'o',mfc='none',label='Atom (simulation)', color='blue', markersize=12)
a1.annotate('$^{87}\mathrm{Rb}$,  2023', (1.02*rb_us[0], rb_us[1]), fontsize=28)
a1.tick_params(axis='both', which='major', labelsize=28)
a1.legend(loc = 'best', fontsize=30)
a1.set_yscale('log')
a1.set_ylabel('Repetition Rate / MHz', fontsize=28)
a1.set_xlabel('Brightness', fontsize=28)
a1.grid( linestyle='-', linewidth=0.2)
f.tight_layout()
f.savefig('pdf_files/source_benchmark.pdf')
```
